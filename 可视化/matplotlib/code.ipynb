```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm

rng = np.random.default_rng(100)
```

```python
# c用于设置颜色，s用于设置marker的大小
data = pd.DataFrame()
data['a'] = np.arange(50)
data['b'] = data['a']**0.5 + 5 * rng.random(50)
data['c'] = rng.integers(0, 50, 50)
data['s'] = rng.normal(1, 1, 50) * 300 + 50

fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter('a', 'b', c='c', s='s', data=data, alpha=0.7)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

```python
# facecolor和edgecolor
# 常用颜色'C0'，边框用黑色'k'即可
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter('a', 'b', data=data, s=400, facecolor='C0', edgecolor='k', alpha=0.7)
```

```python
# hist柱状图
# grid开启格点
# bins柱子的数目或是分界点位置
# density采用概率质量函数表示（放缩）
mu, sigma = 115, 15
x = mu + sigma * rng.standard_normal(10000)
fig, ax = plt.subplots(figsize=(8, 5))

n, bins, patches = ax.hist(x, bins=50, density=True, facecolor='C0', alpha=0.75)

ax.set_xlabel('Length [cm]')
ax.set_ylabel('Probability')
ax.set_title('Aardvark lengths\n (not really)')
ax.text(75, .025, r'$\mu=115,\ \sigma=15$')
ax.axis([55, 175, 0, 0.03])
ax.grid(True)
```

```python
# hist柱状图
# orientation表示柱状图的方向
mu, sigma = 115, 15
x = mu + sigma * rng.standard_normal(10000)
fig, ax = plt.subplots(figsize=(5, 8))

n, bins, patches = ax.hist(x, bins=50, density=True, orientation='horizontal', facecolor='C0', alpha=0.75)

ax.set_xlabel('Probability')
ax.set_ylabel('Length [cm]')
ax.set_title('Aardvark lengths')
ax.grid(True)
```

```python
# colorbar
# extend='both' 会在两端延申尖角
X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
fig, ax = plt.subplots(figsize=(8, 6))
pc = ax.scatter(X, Y, c=Z, cmap='RdBu_r')
fig.colorbar(pc, ax=ax)
```

```python
# colorbar热力图图示
# extend='both' 会在两端延申尖角
# 保存图片 fig.savefig('Figure.png', dpi=600, bbox_inches='tight', transparent=True)
X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
fig, ax = plt.subplots(figsize=(8, 6))
pc = ax.scatter(X, Y, c=Z, cmap=cm.RdBu_r)
fig.colorbar(pc, ax=ax, extend='both')
fig.savefig('Figure.png', dpi=600, bbox_inches='tight', transparent=True)
```

```python
# pcolormesh伪彩色图(网格图)
# vmin, vmax用于colorbar中颜色的标识（会成为端点显示出来）
X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
fig, ax = plt.subplots(figsize=(8, 6))
pc = ax.pcolormesh(X, Y, Z, vmin=-1, vmax=1, cmap='RdBu_r')
fig.colorbar(pc, ax=ax)
```

```python
# contourf填充轮廓线图
# level可以是轮廓线的数目，也可以根据Z值指定
X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
fig, ax = plt.subplots(figsize=(8, 6))
pc = ax.contourf(X, Y, Z, levels=np.linspace(-1.25, 1.25, 11))
fig.colorbar(pc, ax=ax)
```

```python
# contourf填充轮廓线图
# level可以是轮廓线的数目，也可以根据Z值指定
X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
fig, ax = plt.subplots(figsize=(8, 6))
pc = ax.contourf(X, Y, Z, levels=7, cmap=cm.YlGnBu)
fig.colorbar(pc, ax=ax)
```

```python
# bar条形图
# width为条形图宽度0-1
# 设置bottom可以使其堆叠起来
# yerr显示标准差
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]
width = 0.5      # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots(figsize=(8, 6))

ax.bar(labels, men_means, width, yerr=men_std, color=cm.Set2(0), label='Men')
ax.bar(labels, women_means, width, yerr=women_std, color=cm.Set2(2), bottom=men_means,
       label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
```

```python
# bar也可以用于极坐标系
# plt.subplot(projection='polar')
N = 20
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * rng.random(N) + 3
width = 2*np.pi/N * 0.8
colors = cm.viridis(radii / 10.)

ax = plt.subplot(projection='polar')
ax.bar(theta, radii, width=width, bottom=0.0, color=colors, alpha=0.6)
plt.show()
```
