```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
from numpy.polynomial import Polynomial as P 

rng = np.random.default_rng(1)
x = np.linspace(0, 2*np.pi, 20, endpoint=True)
y = np.sin(x) + rng.normal(loc=0.0, scale=0.5, size=x.shape)
plt.scatter(x, y)

poly = P.fit(x, y, deg=5)
print(poly.coef)

xx = np.linspace(0, 2*np.pi, 200)
plt.plot(xx, poly(xx), '--y')

def f(x, a, b, c):
    return a*np.log(x + b) + c*np.arctan(x)
    # return a*x**2 + b*x + c


res = optimize.curve_fit(f, x, y)
print(res)
plt.plot(xx, f(xx, *res[0]), '--g')
```

```python
import numpy as np
from scipy import linalg
rng = np.random.default_rng(1)
A = rng.random((3, 3))
print(A)
A_inv = linalg.inv(A)
# np.allclose(A @ Ainv, np.eye(3))
u, s, v = linalg.svd(A)
# print(u, s, v, sep='\n')
# np.diag(s)
# np.allclose(A, u @ np.diag(s) @ v)
q, r = linalg.qr(A)
print(q, r, sep='\n')
# q.T @ q
np.diag(A)
```

```python
rng = np.random.default_rng(1)
A = rng.random((3, 3))
print(A)
D, P = linalg.eig(A)
# linalg.eigvals(A)
AA = P @ np.diag(D) @ linalg.inv(P)
print(AA)
print(np.allclose(A, AA))
```

```python
import numpy as np
from scipy import linalg
A = np.array([[1, 2], [3, 4]])
la, v = linalg.eig(A)
l1, l2 = la
print(l1, l2)   # eigenvalues
print(v[:, 0])   # first eigenvector
print(v[:, 1])   # second eigenvector
print(np.sum(abs(v**2), axis=0))  # eigenvectors are unitary
v1 = np.array(v[:, 0]).T
print(linalg.norm(A.dot(v1) - l1*v1))  # check the computation
np.allclose(A @ v1, l1 * v1)
```

## 数值微积分

```python
import numpy as np
from scipy import integrate

def f1(t, x, n: int):
    return np.exp(-x*t) / t**n
def f2(x, n):
    return integrate.quad(f1, 1, np.inf, args=(x, n))[0]
n = 100
integrate.quad(f2, 0, np.inf, args=(n,)) # 数值 和 绝对误差
```

```python
import numpy as np
from scipy import integrate
def I(n):
    return integrate.dblquad(lambda t, x: np.exp(-x*t)/t**n, 0, np.inf, lambda x: 1, lambda x: np.inf) # 第一个lambda里面的函数有积分先后顺序的
I(3)
```

## 微分方程

```python
"""ODE数值解
"""

"""demo
求解 theta''(t) + b*theta'(t) + c*sin(theta(t)) = 0
b = 0.25
c = 5.0
要令 omega(t) = theta'(t)

于是
theta'(t) = omega(t)
omega'(t) = -b*omega(t) - c*sin(theta(t))
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def func(y, t, b, c):
    theta, omega = y
    dydt = [omega, -b*omega - c*np.sin(theta)] # 均为一阶
    return dydt
y0 = [np.pi - 0.1, 0.0] # 初值
b = 0.25
c = 5.0
t = np.linspace(0, 10, 300)
res = odeint(func, y0, t, args=(b, c))

plt.plot(t, res[:, 0], '-', label=r'$\theta(t)$')
plt.plot(t, res[:, 1], '-', label=r'$\omega(t)$')
plt.legend()
plt.xlabel('t')
plt.grid()
plt.show()

print(res)
```
