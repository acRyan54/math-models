## MLP多分类

```python
import numpy as np
from sklearn.neural_network import MLPClassifier
x0 = [[0., 0.], [1., 1.]]
y0 = [2, 22]
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(5, 2, 2), random_state=1) # 固定随机种子，可复现

clf.fit(x0, y0)
print(clf.activation)
# clf.alpha
# clf.batch_size
print(clf.coefs_)
xx = [[2, 2], [1, 2]]

print(clf.predict(xx)) # 
print(clf.predict_proba(xx)) # 可以介绍下softmax
# softmax分类 https://blog.csdn.net/lz_peter/article/details/84574716
# 交叉熵损失函数 https://zhuanlan.zhihu.com/p/35709485
```

## 感知机

```python
"""感知机 或
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron

x0 = np.array([[1, 1],[1, 0], [0, 1], [0, 0]])
y0 = np.array([1, 1, 1, 0])
clf = Perceptron(random_state=2).fit(x0, y0)


rng = np.random.default_rng(1)
xx, yy = rng.random((2, 200))
# plt.scatter(xx, yy)
for x, y in tuple(zip(xx, yy)):
    if clf.predict([[x, y]]) == 0:
        plt.scatter(x, y, color='red')
    elif clf.predict([[x, y]]) == 1:
        plt.scatter(x, y, color='blue')
plt.show()

# print(clf.predict([[.5, .5]])) # 只能输出预测的label
```

## $\bigstar$ MLP分类

```python
"""MLP 或
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier

x0 = np.array([[1, 1],[1, 0], [0, 1], [0, 0]])
y0 = np.array([1, 1, 1, 0])
clf = MLPClassifier(hidden_layer_sizes=(5, 5), activation='logistic', solver='lbfgs', random_state=1).fit(x0, y0)


rng = np.random.default_rng(1)
xx, yy = rng.random((2, 200))
# plt.scatter(xx, yy)
for x, y in tuple(zip(xx, yy)):
    if clf.predict([[x, y]]) == 0:
        plt.scatter(x, y, color='red')
    elif clf.predict([[x, y]]) == 1:
        plt.scatter(x, y, color='blue')
plt.show()
```

## $\bigstar$ MLP回归

```python
"""MLP 预测 - 效果并不太好,应该用LSTM或其他时间序列分析算法
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

def func(x):
    return 3*np.log(x+1) + 0.5*np.cos(2*x) - x**(.4)
xx = np.linspace(0, 8, 200)
yy = func(xx)
plt.plot(xx, yy, '-b')

rng = np.random.default_rng(5)
x0 = np.linspace(0, 6, 50)
y0 = func(x0) + rng.normal(loc=0.0, scale=0.2, size=x0.shape)
plt.scatter(x0, y0)

regr = MLPRegressor(hidden_layer_sizes=np.full(10, 10), activation='tanh', random_state=1, max_iter=5000).fit(x0.reshape((-1, 1)), y0)
# regr = MLPRegressor(hidden_layer_sizes=np.full(10, 5), activation='relu', random_state=1, max_iter=5000).fit(x0.reshape((-1, 1)), y0)
# regr = MLPRegressor(hidden_layer_sizes=np.full(10, 5), activation='logistic', random_state=1, max_iter=5000).fit(x0.reshape((-1, 1)), y0)

x1 = np.linspace(0, 8, 200)
# x1 = np.linspace(6, 7, 50)
y1 = regr.predict(x1.reshape(-1, 1))
plt.plot(x1, y1, '-r')
```
